```python
import numpy as np

def rmse(y_true, y_pred):
    """
    Compute Root Mean Square Error (RMSE) between true and predicted values.
    
    Parameters:
    y_true (array): Actual SOH values
    y_pred (array): Predicted SOH values (Kalman-filtered output)
    
    Returns:
    float: RMSE value
    """
    return np.sqrt(np.mean((np.array(y_true) - np.array(y_pred))**2))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from scipy.optimize import differential_evolution

# ------------------------------------------
# ðŸ“¥ Load Data
# ------------------------------------------
df = pd.read_csv("B0005_ML_predictions.csv")

# Use best-performing ML model (Random Forest) as input to Kalman
df['SOH_ML'] = df['RandomForest_SOH_Pred']

# True and ML predicted SOH
true_soh = df['SOH (%)'].values
ml_soh = df['SOH_ML'].values

# ------------------------------------------
# ðŸ§® RMSE Function
# ------------------------------------------
def rmse(y_true, y_pred):
    return np.sqrt(np.mean((np.array(y_true) - np.array(y_pred))**2))

# ------------------------------------------
# ðŸ”„ Kalman Filter
# ------------------------------------------
def apply_kalman_filter(Q, R, P0, measurements):
    x = measurements[0]  # initial estimate
    P = P0
    estimates = [x]

    for z in measurements[1:]:
        # Prediction
        x_pred = x
        P_pred = P + Q

        # Update
        K = P_pred / (P_pred + R)
        x = x_pred + K * (z - x_pred)
        P = (1 - K) * P_pred

        estimates.append(x)

    return np.array(estimates)

# ------------------------------------------
# ðŸŽ¯ Fitness Function for GA
# ------------------------------------------
def fitness(params):
    Q, R, P0 = params
    kalman_soh = apply_kalman_filter(Q, R, P0, ml_soh)
    return rmse(true_soh, kalman_soh)

# ------------------------------------------
# ðŸ§¬ Genetic Algorithm Optimization
# ------------------------------------------
bounds = [
    (1e-6, 1e-1),  # Q
    (1e-6, 1e-1),  # R
    (1, 100)       # P0
]

result = differential_evolution(fitness, bounds, strategy='best1bin', popsize=15, maxiter=100, tol=1e-6)

best_Q, best_R, best_P0 = result.x
print(f"ðŸ§¬ Best Parameters Found:\nQ = {best_Q:.8f}, R = {best_R:.8f}, P0 = {best_P0:.2f}")

# ------------------------------------------
# ðŸ“ˆ Apply Optimized Kalman Filter
# ------------------------------------------
optimized_soh = apply_kalman_filter(best_Q, best_R, best_P0, ml_soh)

# Add to DataFrame
df['SOH_Kalman_Optimized'] = optimized_soh

# Save the result
df.to_csv("B0005_Hybrid_Kalman_Output.csv", index=False)
print("âœ… Output saved to 'B0005_Hybrid_Kalman_Output.csv'")

# ------------------------------------------
# ðŸ“Š Visualization
# ------------------------------------------
plt.figure(figsize=(12, 6))
plt.plot(true_soh, 'k--', label='True SOH')
plt.plot(ml_soh, 'b-', label='ML Predicted SOH (Random Forest)')
plt.plot(optimized_soh, 'orange', label='Kalman Filtered SOH (Optimized)')
plt.xlabel('Cycle Index')
plt.ylabel('SOH (%)')
plt.title('Hybrid Battery Degradation Estimation\nML + Kalman Filter Optimized by Genetic Algorithm')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Save to CSV
output_file = "B0005_Kalman_GA_Optimized.csv"
df.to_csv(output_file, index=False)

print(f"âœ… File saved successfully as: {output_file}")
```
